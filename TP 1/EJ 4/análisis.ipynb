```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("salida.csv")
df
```

```python
total = np.sum(df["Ocurrencia"])
fig, ax = plt.subplots(figsize=(10,6))
bars = plt.bar(df["Base"], df["Ocurrencia"])
ax.bar_label(bars, color='black')
plt.ylabel("Nro de ocurrencias")
plt.xlabel("Base")
plt.show()

```

```python
df["Porcentaje"] = ((df["Ocurrencia"] / total) * 100).round(2)
df
```
