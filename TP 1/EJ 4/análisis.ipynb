```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("salida.csv")
df
```

```python
total = np.sum(df["Ocurrencia"])
fig, ax = plt.subplots(figsize=(10,6))
bars = plt.bar(df["Base"], df["Ocurrencia"])
ax.bar_label(bars, color='black')
plt.ylabel("Nro de ocurrencias")
plt.title("Numero de ocurrencias por base")
plt.xlabel("Base")
plt.savefig("salida/fig1ej4.png")
plt.show()
```

```python
df["Porcentaje"] = ((df["Ocurrencia"] / total) * 100).round(2)
df
```

```python
fig, ax = plt.subplots(figsize=(10,6))
bars = plt.bar(df["Base"], df["Porcentaje"])
ax.bar_label(bars, color='black')
plt.ylabel("Porcentaje")
plt.title("Porcentaje de aparcin por base")
plt.xlabel("Base")
plt.savefig("salida/fig2ej4.png")
plt.show()
```
